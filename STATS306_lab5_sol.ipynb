```
flights %>%
group_by(dest) %>%
summarise(count_flight = n(), mean_dist = mean(distance), mean_arr_delay = mean(arr_delay)) %>%
filter(count_flight > 20, dest != "ALB")
```

```
filter(summarise(group_by(flights, dest), count = n(), mean_dist = mean(distance), mean_arr_delay = mean(arr_delay)), count > 20, dest != "ALB")
```

```
flights %>%
group_by(dest) %>%
filter(!is.na(arr_delay)) %>%
summarise(count_flight = n(), mean_dist = mean(distance), mean_arr_delay = mean(arr_delay))
```
