## **Exercise1**

1.  Explain why each of these strings don’t match a `\`: `"\"`, `"\\"`, `"\\\"`.

-   `"\"`: This will escape the next character in the R string.
-   `"\\"`: This will resolve to `\` in the regular expression, which will escape the next character in the regular expression.
-   `"\\\"`: The first two backslashes will resolve to a literal backslash in the regular expression, the third will escape the next character. So in the regular expression, this will escape some escaped character.


2.  How would you match the sequence `"'\` ?

```r
x <- "\"'\\"
regex <- "^\"'\\\\$"
str_detect(x, regex)
```

3.  How would you match the sequence `"'?\` ?

```r
x <- "\"'?\\"
regex <- "^\"'\\?\\\\$"
str_detect(x, regex)
```

## **Exercise2:**
1.  How would you match the literal string "$^$"?

```r
s <- "$^$"
str_detect(s, "^\\$\\^\\$$")
```

2.  Given the corpus of common words in stringr::words, create regular expressions that find all words that:
   1.   Start with “y”.

```r
str_view(stringr::words, "^y", match = TRUE)
```

   2.   End with “x”

```r
str_view(stringr::words, "x$", match = TRUE)
```

   3.   Are exactly three letters long. (Don’t cheat by using str_length()!)

```r
str_view(stringr::words, "^...$", match = TRUE)
```

   4.   Have seven letters or more.

```r
str_view(stringr::words, "^.{7,}$", match = TRUE)
```

## **Exercise3:**
Construct regular expressions to match words that:
1.   Start and end with the same character(including words with only 1 character).

```r
str_view(stringr::words, "^(.)((.*\\1$)|\\1?)$",match = TRUE)
```

2.   Contain only consonants.

```r
str_view(stringr::words, "^[^aeiou]+$",match = TRUE)
```

3.  Contain two or more vowel-consonant pairs in a row.

```r
stringr::words[str_detect(stringr::words, "([aeiou][^aeiou]){2,}")]
```

4.  Contain a repeated pair of letters (e.g. ``church'' contains ``ch'' repeated twice.)

```r
str_subset(stringr::words, "([A-Za-z][A-Za-z]).*\\1")
```
