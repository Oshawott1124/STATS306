```
library (tidyverse)
```

```
url <- paste0("https://raw.githubusercontent.com/yifanjin328/STATS306/",
              "master/EconomistData.csv")
dat <- read_csv(url)
```

```
head(dat)
```

```
# Exercise 1.1 Create a scatter plot with percent of people over the age of 15 with a bank 
p <- ggplot(
  dat, aes(x = Percent.of.15plus.with.bank.account, y = SEDA.Current.level)) 
p + geom_point()
```

```
# Exercise 1.2 Color the points in the previous plot blue.
p + geom_point(color = "blue")
```

```
# Exercise 1.3 Color the points in the previous plot according to the `Region`.
(p3 <- p + geom_point(aes(color = Region)))
```

```
# Exercise 1.4 Overlay a smoothing line on top of the scatter plot using the default method.
p3 + geom_smooth()
```

```
# Exercise 1.4 Changing the span parameter
p3 + geom_smooth(span = 0.2)
```

```
# Exercise 1.5 Overlay a smoothing line on top of the scatter plot using the lm method
(p5 <- p3 + geom_smooth(method = "lm"))
```

```
# Exercise 1.6 Facetting plots
p5 + facet_wrap(~ Region)
```

```
# Exercise 2.1 Bar plot
barplot <- ggplot(dat, aes(x = Region)) +
  geom_bar() +
  theme(axis.text.x = element_text(angle = 15, hjust = 1))
  barplot
```

```
# Exercise 2.2 Rotate the plot so the bars are horizontal.
barplot + coord_flip()
```

```
# Exercise 3.1 Create boxplots of SEDA scores, `SEDA.Current.level` separately for each `Region`.
plt <- ggplot(dat, aes(x = Region, y = SEDA.Current.level)) + 
  theme(axis.text.x = element_text(angle = 15, hjust = 1))
plt + geom_boxplot() 
```

```
# Exercise 3.2 Overlay points on top of the box plots
plt + geom_boxplot() + geom_point()
```

```
# Exercise 3.3 Jitter each point by a little bit in the horizontal direction.
plt + geom_boxplot() + geom_point(position = position_jitter(w = 0.1))
```

```
plt + geom_boxplot() + geom_jitter(width = 0.1)
```
